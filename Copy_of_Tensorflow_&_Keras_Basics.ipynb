7. Mathematical Operations using Tensor
* How a single neuron return o/p using mathematical approch

```python
X = tf.constant([[2],
                 [3],
                 [4],
                 [5],
                 [6],
                 [7]], dtype = 'float16')
```

```python
Weights = tf.Variable([[0.1],
                       [0.3],
                       [0.2],
                       [0.4],
                       [0.6],
                       [0.5]], dtype = 'float16')
```

```python
WT = tf.transpose(Weights)
```

```python
WT
```

```python
# g(x) = WT * X + bias
gx = tf.matmul(WT, X)
```

```python
b0 = tf.constant(1.0, dtype = 'float16')
w0 = tf.Variable(1.7, dtype = 'float16')
# calculate b = b0 * w0
b = tf.multiply(b0, w0)
```

```python
g_x = tf.matmul(WT, X) + b
```

```python
g_x
```

```python
Y = tf.sigmoid(g_x)
```

```python
Y
```

```python
Y = tf.nn.relu(g_x)
Y
```

8. Dense Neural Network - Multiclass Classification

```python
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam, Adagrad
```

```python
def build_model():
    model = Sequential()
    # Input Shape = Number of Features or Shape of Features
    model.add(Input(shape = (8,)))
    # Input Layer
    # units = 32 (Number of Neurons), Activation Function = 'relu' or 'leakyrelu'
    model.add(Dense(units = 32, activation = 'relu'))
    # Hidden Layer - I
    model.add(Dense(units = 64, activation = 'relu'))
    # Hidden Layer - II
    model.add(Dense(units = 128, activation = 'relu'))
    # Output Layer
    model.add(Dense(units = 3, activation = 'softmax'))
    # Compile the model
    model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])

    return model
```

```python
model = build_model()
```

```python
model.summary()
```
