##### Pytorch:
* PyTorch is an open-source deep learning framework developed by Facebook’s AI Research lab (FAIR). It is widely used for machine learning, deep learning, and AI applications due to its ease of use, flexibility, and dynamic computation graph.

* It supports multi-dimentional array (Tensors) with GPU acceleration.
* Optimized for GPUs: PyTorch allows seamless execution on CPU and GPU.
* Community Support: Strong ecosystem with pre-trained models (Torchvision, GenAI - Hugging Face)

```python
import torch
import numpy as np
```

1. Scalar (Rank 0 Tensor)
  * torch.tensor() : Create a tensor data of value contant type.
  * torch.tensor(data, dtype = None, device = None, requires_grad = False)

```python
# Data = Input Data (list, numpy array, or any other tensor shape)
# dtype = Specify the data type (e.g. torch.float32, torch.float64 etc.)
# device = 'None', 'cuda', 'cpu'
const_1 = torch.tensor(10)
```

```python
const_1
```

2. Vector (Rank 1 Tensor)

```python
const_2 = torch.tensor([10, 20, 30, 40, 50, 60, 70, 80], dtype = torch.int16, device = 'cuda')
```

```python
print(const_2)
```

```python
# get Rank (Dimentionality)
const_2.dim()
```

3. Matrix (Rank 2)

```python
# Rank 2 Tensor
const_3 = torch.tensor([[11, 12, 13, 14, 15, 16, 17],
                        [21, 22, 23, 24, 25, 26, 27],
                        [31, 32, 33, 34, 35, 36, 37]], dtype = torch.int16, device = 'cuda')
```

```python
print(const_3)
```

```python
# get Rank (Dimentionality)
const_3.dim()
```

4. Tensor (Rank 3 or More)

```python
const_4 = torch.tensor([[[11, 12, 13, 14, 15, 16, 17],
                        [21, 22, 23, 24, 25, 26, 27],
                        [31, 32, 33, 34, 35, 36, 37]],

                        [[11, 12, 13, 14, 15, 16, 17],
                        [21, 22, 23, 24, 25, 26, 27],
                        [31, 32, 33, 34, 35, 36, 37]],

                        [[11, 12, 13, 14, 15, 16, 17],
                        [21, 22, 23, 24, 25, 26, 27],
                        [31, 32, 33, 34, 35, 36, 37]]], dtype = torch.int16, device = 'cuda')
```

```python
print(const_4)
```

Tensor : A tensor is a multi-dimensional array. It generalizes as scalar, vectors, matrices and higher dimensions.

```python
mat_1 = np.random.randint(0, 255, size = (10,10))
```

```python
# uint8 - Unsigned Integer 8, used for positive value only
const_5 = torch.tensor(mat_1, dtype = torch.uint8)
```

5. Display as GrayScale - 2D Image

```python
import matplotlib.pyplot as plt
plt.imshow(const_5, cmap = 'gray')
```

```python
mat_1 = np.random.randint(0, 255, size = (10,10))
mat_2 = np.random.randint(0, 255, size = (10,10))
mat_3 = np.random.randint(0, 255, size = (10,10))
```

```python
# uint8 - Unsigned Integer 8, used for positive value only
const_6 = torch.tensor([mat_1, mat_2, mat_3], dtype = torch.uint8)
```

```python
# reshape tensor from (3, 10, 10) to (10, 10, 3)
const_7 = const_6.permute(1, 2, 0)
```

```python
print(const_7.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(const_7)
```
