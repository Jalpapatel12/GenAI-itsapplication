7. Mathematical Operations using Tensor
    
  * How a single neuron return o/p using mathematical approach.

```python
X = torch.tensor([[2],
                  [3],
                  [4],
                  [5],
                  [6],
                  [7]], dtype = torch.float16)
```

```python
Weights = torch.tensor([[0.1],
                        [0.3],
                        [0.2],
                        [0.5],
                        [0.6],
                        [0.2]], dtype = torch.float16, requires_grad = True)
```

```python
WT = Weights.T
```

```python
# Compute Output (WT * X)
# '@' is matrix multiplication in PyTorch
g_x = WT @ X
```

```python
print(g_x)
```

```python
b0 = torch.tensor(1.0, dtype = torch.float16)
w0 = torch.tensor(1.5, dtype = torch.float16)
```

```python
b = torch.multiply(b0, w0)
```

```python
g_x = torch.matmul(WT, X) + b
```

```python
g_x
```

```python
torch.relu(g_x)
```

```python
Y = torch.sigmoid(g_x)
```

```python
# detach() - Detaches the tensor from tensor computation graph - creates new Tensor that shares the same data
print("Linear Neuron Output:", g_x.detach().numpy())
print("Activated Output:", Y.detach().numpy())
```

```python
Y = torch.relu(g_x)
```

```python
print("Linear Neuron Output:", g_x.detach().numpy())
print("Activated Output:", Y.detach().numpy())
```

8. Dense Neural Network - Multiclass Classification

```python
import torch
import torch.nn as nn
from torch.optim import Adam
```

```python
# Define Model
model = nn.Sequential(
    # Input Layer
    nn.Linear(in_features = 8, out_features = 32, bias = True),
    nn.ReLU(),
    # Hidden Layer - I
    nn.Linear(in_features = 32, out_features = 64, bias = True),
    nn.ReLU(),
    # Hidden Layer - II
    nn.Linear(in_features = 64, out_features = 128, bias = True),
    nn.ReLU(),
    # Output Layer
    nn.Linear(in_features = 128, out_features = 3, bias = True),
    nn.Softmax(dim = 1)
)

# Define Loss Function and Model Optimizer
criterion = nn.CrossEntropyLoss()
# Model Optimizers = Weights & Biases, Learning_rate = (0.0001 - 0.1)
optimizer = Adam(model.parameters(), lr = 0.001)
# Print Model Summary
print(model)
```
